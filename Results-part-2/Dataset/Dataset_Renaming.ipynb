```python
import os
import cv2
import imutils
```

```python
def crop_images(dataset):
	for im_num in range(20, 20 + len(os.listdir(dataset))):
		# file_name = f'image{im_num}.png'
		# file_path = os.path.join(dataset, 'image')
		im_arr = cv2.imread(dataset + f'/image{im_num}.png', 0)
		H, W = im_arr.shape
		dH = 960-H
		dW = 1280-W
		if(dH<0 or dW<0):
			im_arr = im_arr[:960, :1280]
		else:
			im_arr = cv2.copyMakeBorder(im_arr, 0, dH, 0, dW, cv2.BORDER_REFLECT)
		cv2.imwrite(dataset + f'/image{im_num}.png', im_arr)
```
